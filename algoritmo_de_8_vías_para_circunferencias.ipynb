```
import numpy as np
from matplotlib.pyplot import imshow
from PIL import Image ,ImageDraw
import matplotlib.pyplot as plt

def ROUND(a):
	return int(a + 0.5)
 
 

#modulo de 8 vias
def circulo8(cx,cy,r):
  img_draw.point((cx,cy+r),fill='black')
  img_draw.point((cx,cy-r),fill='black')
  img_draw.point((cx+r,cy),fill='black')
  img_draw.point((cx-r,cy),fill='black')
  x=1
  y=ROUND(np.sqrt(r**2-x**2))
  while x<y:
    img_draw.point((cx+x,cy+y),fill='blue')
    img_draw.point((cx+x,cy-y),fill='blue')
    img_draw.point((cx-x,cy+y),fill='blue')
    img_draw.point((cx-x,cy-y),fill='blue')
    img_draw.point((cx+y,cy+x),fill='blue')
    img_draw.point((cx+y,cy-x),fill='blue')
    img_draw.point((cx-y,cy+x),fill='blue')
    img_draw.point((cx-y,cy-x),fill='blue')
    x+=1
    y=ROUND(np.sqrt(r**2-x**2))

  if x==y:
    img_draw.point((cx+x,cy+y),fill='blue')
    img_draw.point((cx+x,cy-y),fill='blue')
    img_draw.point((cx-x,cy+y),fill='blue')
    img_draw.point((cx-x,cy-y),fill='blue')



#diibujar camvas
blank_image=Image.new('RGB', (100,100), 'white')
img_draw=ImageDraw.Draw(blank_image)


circulo8(50,50,40)


blank_image.save('drawn_image.jpg')
# Open your file image using the path
img = Image.open('drawn_image.jpg')

# Since plt knows how to handle instance of the Image class, just input your loaded image to imshow method
plt.imshow(img,origin='lower')
```
